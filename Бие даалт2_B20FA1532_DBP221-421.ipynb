```python
# 8. Тухайн хичээлд “A”, ”B”, ”C”, ”D”, ”F” дүн авсан оюутнуудын тоог ашиглан Piechart үүсгэнэ үү.


b['Үнэлгээ']=np.where((b['Нийт'] < 59.5 ),'F',
                  np.where((b['Нийт'] >= 59.5 )&(b['Нийт'] < 69.5 ),'D',
                  np.where((b['Нийт'] >= 69.5 )&(b['Нийт'] < 79.5 ),'C',
                  np.where((b['Нийт'] >= 79.5 )&(b['Нийт'] < 89.5 ),'B',
                  np.where((b['Нийт'] >= 89.5 ),'A','NAN'
                          )))))

c1 = b['Үнэлгээ'].value_counts()['A']
c2 = b['Үнэлгээ'].value_counts()['B']
c3 = b['Үнэлгээ'].value_counts()['C']
c4 = b['Үнэлгээ'].value_counts()['D']
c5 = b['Үнэлгээ'].value_counts()['F']



Үнэлгээ = ['A','B','C','D','F']
data = [c1, c2, c3, c4, c5]

explode = (0.1, 0.0, 0.2, 0.3, 0.0)
colors = ( "orange", "green", "#444444",
           "pink", "beige")

wp = { 'linewidth' : 1, 'edgecolor' : "green" }

def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)
 
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(data,
                                  autopct = lambda pct: func(pct, data),
                                  explode = explode,
                                  labels = Үнэлгээ,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))

ax.legend(wedges, Үнэлгээ,
          title ="Үнэлгээ",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
 
plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Үнэлгээний дугуй диаграмм")
 
plt.show()
```
